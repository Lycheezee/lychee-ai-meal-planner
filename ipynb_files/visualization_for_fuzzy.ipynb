```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# BMI and Age categories
bmi_categories = ['Underweight', 'Normal', 'Overweight', 'Obese']
age_groups = ['Youth', 'Young Adult', 'Adult', 'Middle-Aged', 'Senior', 'Elderly']

# Example adjustment matrices for each exercise level
exercise_levels = {
    'Sedentary': [
        [1.10, 1.05, 1.00, 0.98, 1.00, 1.10],
        [0.95, 0.90, 0.90, 0.88, 0.90, 0.85],
        [0.85, 0.88, 0.85, 0.83, 0.80, 0.78],
        [0.75, 0.78, 0.75, 0.70, 0.70, 0.75],
    ],
    'Light': [
        [1.15, 1.10, 1.05, 1.00, 1.05, 1.05],
        [1.00, 0.98, 0.98, 0.95, 1.00, 0.90],
        [0.92, 0.95, 0.92, 0.90, 0.88, 0.85],
        [0.85, 0.88, 0.85, 0.80, 0.80, 0.82],
    ],
    'Moderate': [
        [1.20, 1.15, 1.10, 1.05, 1.10, 1.10],
        [1.05, 1.00, 1.00, 0.98, 1.00, 0.92],
        [0.95, 0.98, 0.95, 0.92, 0.90, 0.88],
        [0.85, 0.88, 0.88, 0.82, 0.80, 0.82],
    ],
    'Active': [
        [1.25, 1.20, 1.10, 1.05, 1.10, 1.10],
        [1.05, 1.00, 1.00, 0.98, 1.05, 0.92],
        [0.95, 1.00, 1.00, 0.85, 0.92, 0.92],
        [0.80, 0.90, 0.90, 0.70, 0.75, 0.85],
    ],
    'Very Active': [
        [1.30, 1.25, 1.15, 1.10, 1.15, 1.15],
        [1.10, 1.05, 1.05, 1.00, 1.10, 0.95],
        [1.00, 1.05, 1.05, 0.95, 1.00, 0.95],
        [0.85, 0.95, 0.95, 0.85, 0.90, 0.90],
    ],
}
```

```python
# Create heatmaps
fig, axes = plt.subplots(5, 1, figsize=(5, 25), sharey=True)

for ax, (level, data) in zip(axes, exercise_levels.items()):
    df = pd.DataFrame(data, index=bmi_categories, columns=age_groups)
    sns.heatmap(df, annot=True, cmap='coolwarm', ax=ax, cbar=False)
    ax.set_title(level)
    ax.set_xlabel("Age Group")
    ax.set_ylabel("BMI Category")

plt.suptitle("Fuzzy Calorie Adjustment by Exercise Level", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
